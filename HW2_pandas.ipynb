```python
import pandas as pd

# create a dataframe

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 17, 28],
    'City': ['New York', 'San Francisco', 'Los Angeles', 'Alexandria'],
    'Salary': [100, 150, 200, 300]
}

df = pd.DataFrame(data)
df
```

```python
df['Name']
df[['Name','Salary']]
```

```python
df.iloc[0]
df.iloc[2]
```

```python
# calculate some statistics on each column

df['Age'].mean()
```

```python
# calculate some statistics on each column

print("The minimum salary is", df['Salary'].min())
print("The maximum salary is", df['Salary'].max())
print("The mean salary is", df['Salary'].mean())
```

```python
# calculate some statistics on each column

print("The minimum age is", df['Age'].min())
print("The maximum age is", df['Age'].max())
print("The mean age is", df['Age'].mean())
```

```python
ndata = [10, 20, 30, 40, 50]
s = pd.Series(ndata)

print(s)
print(s.min())
print(s.max())
print(s.mean())
```
